# Exercise 2 - alternative

```python
#ALTERNATIVE - we can also just drop them from the list manually
doi_agent = Agent(MODEL, output_type=list[str], system_prompt="Extract doi numbers from the provided articles")
@doi_agent.output_validator
def validate_output(output: list[str]) -> list[str]:
    valid_dois = []
    for doi in output:
        if is_valid_doi(doi):
            valid_dois.append(doi)
    return valid_dois
result = doi_agent.run_sync(str(PAPERS)) #Needs to be sting-ified
pretty_print_trace(result)
result.output
```
